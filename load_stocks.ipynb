```python
def from_json(file_name) -> dict: #takes in the json file name. Returns a dictionary of stock objects that are indexed by their ticker name
    f = open(file_name, "r")
    ret = {}
    parse = json.loads(f.read())
    for i in parse:
        dates = []
        for d in i["dates"]: #converts all the isoformat dates from the JSON into datetime objects
            dates.append(datetime.date.fromisoformat(d))
        ret[i["ticker"]] = stock(i["ticker"], i["prices"], dates, i["industry"])
    f.close()
    return ret
```
