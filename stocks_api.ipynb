#### from_api(url)
loads data from current API

```python
def from_api(url) -> tuple:
    response = requests.get(url)
    if (response.status_code != 200):
        return 0
    parse = response.json()
    if (parse["status"].lower() != "ok"):
        return 0
    ticker = (parse["meta"]["symbol"]).upper()
    industry = stocks[stocks["Symbol"] == ticker]["GICS Sector"].iloc[0]
    prices = []
    dates = []
    for i in parse["values"]:
        dates.append(datetime.date.fromisoformat(i["datetime"]))
        prices.append(i["close"])
    return (ticker,prices, dates, industry)

def from_batch_api(url) -> int | list[tuple]:
    response = requests.get(url)
    print(url)
    if (response.status_code != 200):
        return 0
    parse = response.json()
    data_ = {}
    for symbol, symbol_data in parse.items():
        if symbol_data["status"].lower() == "ok":
            ticker = symbol.upper()
            stock_entry = stocks[stocks["Symbol"] == ticker].iloc[0]
            industry = stock_entry["GICS Sector"]
            prices = []
            dates = []
            for i in symbol_data["values"]:
                dates.append(datetime.date.fromisoformat(i["datetime"]))
                prices.append(i["close"])
            keywords = [
                word.strip().lower() for col in [
                    stock_entry["Security"],
                    stock_entry["GICS Sector"],
                    stock_entry["GICS Sub-Industry"],
                    str(stock_entry["Headquarters Location"].split(',')[0]),
                    str(stock_entry["Headquarters Location"].split(',')[1])
                ] for word in col.split()
            ]
            data_.append((ticker,prices, dates, industry, keywords))
            
    return data_
```

### JSON methods
#### to_json(file_name, stocks)
Writes stocks as a json string to file name


```python
def to_json(file_name: str, stocks: list[stock]) -> None:   
    with open(file_name, "w") as file:
        json.dump([stock_.to_json() for stock_ in stocks], file)
```

#### from_json(file_name)
Load data from json file into a dictionary of stock objects

```python
def from_json(file_name: str) -> dict: #takes in the json file name. Returns a dictionary of stock objects that are indexed by their ticker name
    with open(file_name, "r") as file:
        data = json.load(file)

    return {
        entry["ticker"]: stock(
            entry["ticker"],
            entry["prices"],
            [datetime.date.fromisoformat(d) for d in entry["dates"]],
            entry["industry"],
            entry["keywords"]
        )
        for entry in data
    }
```
